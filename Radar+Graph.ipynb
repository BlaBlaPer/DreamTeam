```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
from math import pi
```

```python
labels = ['Snipers', '          Assaults', '          Supports','Captain          ', 'Strategist          ']
# spaces are used to alter the position of labels in graph

# for example, set values to:
values = [8.6, 2.0, 6.5, 7.0, 7.3]
#values_new = [10, 10, 4, 7, 8]

N = len(labels)

x_as = [n / float(N) * 2 * pi for n in range(N)]

values += values[:1]
#values_new += values_new[:1]
x_as += x_as[:1]

plt.rc('axes', linewidth=1, edgecolor="#888888")

ax = plt.subplot(111, polar=True)

ax.set_theta_offset(pi / 2)
ax.set_theta_direction(-1)

ax.set_rlabel_position(0)

ax.xaxis.grid(True, color="#888888", linestyle='solid', linewidth=1)
ax.yaxis.grid(True, color="#888888", linestyle='solid', linewidth=1)

plt.xticks(x_as[:-1], labels, size=12, verticalalignment='baseline')
matplotlib.rcParams['xtick.direction'] = 'out'
plt.yticks([0, 2, 4, 6, 8, 10], ["0","2", "4", "6", "8", "10"], size=7)

ax.plot(x_as, values, linewidth=1, linestyle='solid', zorder=3, label='Original')
ax.fill(x_as, values, 'b', alpha=0.2)

#ax.plot(x_as, values_new, linewidth=1, linestyle='solid', zorder=3, label='New')
#ax.fill(x_as, values_new, 'r', alpha=0.2)

ax.legend(loc='best', bbox_to_anchor=(0, 0))

plt.ylim(0, 10)

plt.show()
```

```python
labels = ['Snipers', '          Assaults', '          Supports','Captain          ', 'Strategist          ']
# spaces are used to alter the position of labels in graph

# for example, set values to:
values = [8.6, 2.0, 6.5, 7.0, 7.3]
values_new = [7.9, 7.2, 5.0, 7.0, 7.9]


N = len(labels)

x_as = [n / float(N) * 2 * pi for n in range(N)]

values += values[:1]
values_new += values_new[:1]
x_as += x_as[:1]

plt.rc('axes', linewidth=1, edgecolor="#888888")

ax = plt.subplot(111, polar=True)

ax.set_theta_offset(pi / 2)
ax.set_theta_direction(-1)

ax.set_rlabel_position(0)

ax.xaxis.grid(True, color="#888888", linestyle='solid', linewidth=1)
ax.yaxis.grid(True, color="#888888", linestyle='solid', linewidth=1)

plt.xticks(x_as[:-1], labels, size=12, verticalalignment='baseline')
matplotlib.rcParams['xtick.direction'] = 'out'
plt.yticks([0, 2, 4, 6, 8, 10], ["0","2", "4", "6", "8", "10"], size=7)

ax.plot(x_as, values, linewidth=1, linestyle='solid', zorder=3, label='Original')
ax.fill(x_as, values, 'b', alpha=0.2)

ax.plot(x_as, values_new, linewidth=1, linestyle='solid', zorder=3, label='New')
ax.fill(x_as, values_new, 'r', alpha=0.2)

ax.legend(loc='best', bbox_to_anchor=(0, 0))

plt.ylim(0, 10)

plt.show()
```
